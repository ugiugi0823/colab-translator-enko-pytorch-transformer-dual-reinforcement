# DSL 만들기

```python
CMD = "python dual_train.py"\
    "--model_fn ./model_save/dsl.pth "\
    "--lm_fn ./model_save/test2.pth"\
    "--train ./data/corpus.shuf.train.tok.bpe "\
    "--valid ./data/corpus.shuf.valid.tok.bpe "\
    "--lang enko "\
    "--gpu_id 0 "\
    "--batch_size 8 "\
    "--n_epochs 1 "\
    "--max_length 64 "\
    "--dropout .2 "\
    "--hidden_size 768 "\
    "--n_layers 4 "\
    "--max_grad_norm 1e+8 "\
    "--iteration_per_update 4 "\
    "--dsl_n_warmup_epochs 0"\

!{CMD}
```
